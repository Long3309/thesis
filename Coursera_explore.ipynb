```python
import pandas as pd
course_df = pd.read_csv("data/Coursera.csv")
course_df.head()
```

```python
list(course_df["Course Rating"].unique())
```

```python
def get_unique_skills(skills_list):
    unique_skills = list()
    for row in range(len(skills_list)):
        for skill in skills_list[row].split():
            skill = skill.lower()
            if skill not in unique_skills:
                unique_skills.append(skill)
    return unique_skills
```

```python
unique_skills = get_unique_skills(course_df["Skills"])
len(unique_skills)
```
